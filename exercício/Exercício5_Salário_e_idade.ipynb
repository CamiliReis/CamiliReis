# Exercício 5 — Salário e idade



```python
idade= int(input("Qual sua idade?: "))
print(f"Minha idade é: {idade} anos")
print()

salario= float(input("Qual o valor da sua renda mensal?:"))
print(f"Minha renda mensal é de R$ {salario:.2f}")
print()

print("A pessoa é maior de idade?" , idade>=18)
print("O salario é 2 mil ou acima?" , salario>= 2000)
print("É maior de idade e ganha pelo menos 2 mil?" , idade>= 18 and salario>=2000)
```
